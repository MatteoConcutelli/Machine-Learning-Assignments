```python
import pandas as pd
import json
```

```python
basic_CNN = './Parameter_Search_Results/basic_CNN_results.json'
basic_CNN_edges = './Parameter_Search_Results/basic_CNN_edges_results.json'
basic_CNN_augmented = './Parameter_Search_Results/basic_CNN_augmented_results.json'
basic_CNN_augmented_edges = './Parameter_Search_Results/basic_CNN_augmented_edges_results.json'

# Caricare i dati dal file JSON
with open(basic_CNN, 'r') as file:
    basic_CNN_results = json.load(file)

with open(basic_CNN_edges, 'r') as file:
    basic_CNN_edges_results = json.load(file)

with open(basic_CNN_augmented, 'r') as file:
    basic_CNN_augmented_results = json.load(file)

with open(basic_CNN_augmented_edges, 'r') as file:
    basic_CNN_augmented_edges_results = json.load(file)
    

# Creazione del DataFrame
basic_CNN_df = pd.DataFrame(basic_CNN_results)
basic_CNN_edges_df = pd.DataFrame(basic_CNN_edges_results)
basic_CNN_augmented_df = pd.DataFrame(basic_CNN_augmented_results)
basic_CNN_augmented_edges_df = pd.DataFrame(basic_CNN_augmented_edges_results)

# Mostrare le prime righe per analisi preliminare
basic_CNN_df.head(), basic_CNN_df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
basic_CNN_df = pd.DataFrame(basic_CNN_results)
basic_CNN_edges_df = pd.DataFrame(basic_CNN_edges_results)
basic_CNN_augmented_df = pd.DataFrame(basic_CNN_augmented_results)
basic_CNN_augmented_edges_df = pd.DataFrame(basic_CNN_augmented_edges_results)
```

```python
# Creazione dei grafici con lineplot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# epochs vs Accuracy Rate (senza augmentation)
sns.lineplot(x="epochs", y="final_accuracy", data=basic_CNN_df, ax=axes[0, 0])
axes[0, 0].set_title("Normal Data (No Aug)")
axes[0, 0].set_xlabel("Epochs")
axes[0, 0].set_ylabel("Final Accuracy (%)")

# epochs vs Accuracy Rate (con marked edges senza augmentation)
sns.lineplot(x="epochs", y="final_accuracy", data=basic_CNN_edges_df, ax=axes[0, 1])
axes[0, 1].set_title("Marked Edges Data (No Aug)")
axes[0, 1].set_xlabel("Epochs")
axes[0, 1].set_ylabel("Final Accuracy (%)")

# epochs vs Accuracy Rate (con augmentation)
sns.lineplot(x="epochs", y="final_accuracy", data=basic_CNN_augmented_df, ax=axes[1, 0])
axes[1, 0].set_title("Normal Data Augmented")
axes[1, 0].set_xlabel("Epochs")
axes[1, 0].set_ylabel("Final Accuracy (%)")

# epochs vs Accuracy Rate (con marked edges e augmentation)
sns.lineplot(x="epochs", y="final_accuracy", data=basic_CNN_augmented_edges_df, ax=axes[1, 1])
axes[1, 1].set_title("Marked Edges Data Augmented")
axes[1, 1].set_xlabel("Epochs")
axes[1, 1].set_ylabel("Final Accuracy (%)")

# Adjust layout to make sure there is enough space between subplots
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creazione di un DataFrame unificato con l'etichetta per ciascun dataset
import pandas as pd

# Aggiungiamo un campo per identificare facilmente i vari dataset nei dati
basic_CNN_df['dataset'] = 'Normal Data (No Aug)'
basic_CNN_edges_df['dataset'] = 'Marked Edges Data (No Aug)'
basic_CNN_augmented_df['dataset'] = 'Normal Data Augmented'
basic_CNN_augmented_edges_df['dataset'] = 'Marked Edges Data Augmented'

# Concatenazione dei dataframe
all_datasets_df = pd.concat([basic_CNN_df, basic_CNN_edges_df, 
                             basic_CNN_augmented_df, basic_CNN_augmented_edges_df])

# Creazione del boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x="dataset", y="final_accuracy", data=all_datasets_df)

# Titolo e etichette
plt.title("Comparison of Final Accuracy Across Different Datasets")
plt.xlabel("Dataset")
plt.ylabel("Final Accuracy (%)")

# Mostra il grafico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
